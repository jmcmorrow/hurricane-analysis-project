```python
# write your update damages function here:
def update_damages(damages):
    '''
    Changes strings to floats
    '''
    new_damages = []
    for damage in damages:
        if damage == 'Damages not recorded':
            new_damages.append('Damages not records')
        elif 'M' in damage:
            num = float(damage[:-1])
            new_damages.append(float(num*1000000))
        elif 'B' in damage:
            num = float(damage[:-1])
            new_damages.append(float(num*1000000000))
    return new_damages

new_damages = update_damages(damages)
new_damages
```

```python
# write your construct hurricane dictionary function here:
def construct_hurricane(names, months, years, max_sustained_winds, areas_affected, damages, deaths):
    '''
    Returns a dictionary with all of the information for each hurricane.
    '''
    hurricanes = {}
    for hurricane in range(len(names)):
        new_dict = {'Name': names[hurricane], 
                    'Month': months[hurricane], 
                    'Year': years[hurricane], 
                    'Max Sustained Winds': max_sustained_winds[hurricane], 
                    'Areas Affected': areas_affected[hurricane], 
                    'Damages': damages[hurricane], 
                    'Deaths': deaths[hurricane]}
        hurricanes.update({names[hurricane]: new_dict})
    return hurricanes

hurricane_dictionary = construct_hurricane(names, months, years, max_sustained_winds, areas_affected, new_damages, deaths)
hurricane_dictionary
```

```python
# write your construct hurricane by year dictionary function here:
def hurricane_by_year(hurricanes):
    year = {}
    for hurricane in hurricanes.values():
        year.update({hurricane['Year']: hurricane})
    return year
hurricane_dict_by_year = hurricane_by_year(hurricane_dictionary)
hurricane_dict_by_year
```

```python
# write your count affected areas function here:
def count_affected_areas(hurricanes):
    areas = {}
    for hurricane in hurricanes.values():
        for area in hurricane['Areas Affected']:
            if area not in areas:
                areas.update({area: 1})
            else:
                areas[area] += 1
    return areas
count_areas = count_affected_areas(hurricane_dictionary)
count_areas
```

```python
# write your find most affected area function here:
def most_affected_area(affected_area):
    area, most = (None, 0)
    for key, value in affected_area.items():
        if value > most:
            area, most = (key, value)
    return area

most_affected = most_affected_area(count_areas)
most_affected
```

```python
# write your greatest number of deaths function here:
def most_deaths(hurricanes):
    area, deaths = (None, 0)
    for key, item in hurricanes.items():
        if item['Deaths'] > deaths:
            deaths = item['Deaths']
            area = key
    return area

most_deaths(hurricane_dictionary)
```

```python
# write your catgeorize by mortality function here:
def categorize_by_mortality(hurricanes):
    zero = []
    one = []
    two = []
    three = []
    four = []
    five = []
    for value in hurricanes.values():
        if value['Deaths'] >= 10000:
            five.append(value)
        elif value['Deaths'] >= 1000:
            four.append(value)
        elif value['Deaths'] >= 500:
            three.append(value)
        elif value['Deaths'] >= 100:
            two.append(value)
        elif value['Deaths'] > 0:
            one.append(value)
        else:
            zero.append(value)
    mortality_dict = {0: zero, 1: one, 2: two, 3: three, 4: four, 5: five}
    return mortality_dict

mortality = categorize_by_mortality(hurricane_dictionary)
mortality
```

```python
# write your greatest damage function here:
def greatest_damage(hurricanes):
    area, greatest = (None, 0)
    for key, value in hurricanes.items():
        if value['Damages'] == 'Damages not records':
            continue
        elif value['Damages'] > greatest:
            area = key
            greatest = value['Damages']
    return area, greatest

most_damage = greatest_damage(hurricane_dictionary)
most_damage
```

```python
# write your catgeorize by damage function here:
def rate_by_damage(hurricanes):
    zero = []
    one = []
    two = []
    three = []
    four = []
    five = []
    for value in hurricanes.values():
        if value['Damages'] == 'Damages not records':
            continue
        elif value['Damages'] >= 50000000000:
            five.append(value)
        elif value['Damages'] >= 10000000000:
            four.append(value)
        elif value['Damages'] >= 1000000000:
            three.append(value)
        elif value['Damages'] >= 100000000:
            two.append(value)
        elif value['Damages'] > 0:
            one.append(value)
        else:
            zero.append(value)
    damage_dict = {0: zero, 1: one, 2: two, 3: three, 4: four, 5: five}
    return damage_dict

damage_ranking = rate_by_damage(hurricane_dictionary)
damage_ranking
```
